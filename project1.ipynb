```python
colors = ['#ff033e','#0000FF','#008000','#FFFF00']
cols = ['account_length','total_income','age','years_employed']
for c in cols:
    df[c] = df[c].astype(float)
for i, col in enumerate(cols):
    plt.figure(figsize=(6, 4))
    plt.hist(df[col],bins=10,edgecolor='black',color=colors[i% len('colors')])
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(False)
    plt.show()
    
        
```

```python
plt.figure(figsize=(5,4))
sns.countplot(data=df, x='gender',hue = 'gender', palette=sns.color_palette("cool", 2),legend=True)
plt.title('Gender Distribution (0=Male, 1 = Female')
plt.show()
print("\n If the count plot shows 0(Male) than 1 (Female),it means the dataset contains more male customers than female customers..")
print("\n This might affect how products or service are targeted to different genders..")
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,4))
sns.countplot(data=df, x="own_car", ax=ax[0], hue='own_car', palette='coolwarm',legend=True)
ax[0].set_title("Car Ownership (0='No',1='Yes')")

sns.countplot(data=df, x='own_property', ax=ax[1], hue='own_property', palette='coolwarm', legend=True)
ax[1].set_title('Property Ownership (0=No, 1=Yes)')

plt.show()

print("If most values are 0, it means most people dont own a car.")
print("If there are more 1s, then a large portion of the population owns cars  indicating better income or living standard.")


print("A higher number of 1s shows that many individuals own property (house/land).This can be linked to financial stability.")
```

```python
cols = ['work_phone','phone','email']
df[cols].sum().plot(kind='bar',color = ['skyblue','blue','green'],legend=True)
plt.title("Commnication access tool..")
plt.ylabel("Number of People")
plt.show()

```

```python
# Family and Children Distribution
fig, ax = plt.subplots(1,2, figsize=(10,4))

sns.histplot(df['num_children'], bins=10, kde=True, ax=ax[0], color = 'green', legend=True)
ax[0].set_title('Number of Children')

sns.histplot(df['num_family'], bins=10, kde=True, ax = ax[1], color = 'orange', legend=True)
ax[1].set_title("Number of Family Members..")
plt.show()
print('Histogram shows how many people have 0, 1, 2, or more children.A right-skewed graph (many 0s and few large numbers) means most people have few or no children.')

print("Shows household size  large families may indicate traditional or joint living systems, while smaller families may show modern nuclear households.")
```

```python
#Unemployment and Gender Relation
sns.countplot(data=df, x='gender', hue='unemployed', palette="Set1")
plt.title("Umemployment Status by Gender")
plt.show()

print("If Unemployed = 1 counts are high, that means a large portion of individuals are unemployed.Analyzing this with gender can show whether unemployment is more common among males or females.")
```

```python
#Credit Card Eligibility (Target Variable)
plt.figure(figsize=(5,4))
sns.countplot(data=df, x='target', hue='target', palette='Set2')
plt.title('Credit Card Eligibility (0= Not Eligible, 1= Eligible)')
plt.show()
print('Shows how many people are eligible or not for a credit card.If majority are 0, few people qualify; if 1 dominates, most are eligible.')
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 7))

sns.countplot(data=df, x='income_type', ax=ax[0], hue='income_type', palette='Set3',legend=True)
ax[0].set_title("Income Type Distribution")
ax[0].tick_params(axis='x', rotation=45)

sns.countplot(data=df, x='education_type', ax=ax[1],hue='education_type', palette='Set2',legend=True)
ax[1].set_title("Education Level Distribution")
ax[1].tick_params(axis='x', rotation=45)

sns.countplot(data=df, x='family_status', ax=ax[2], hue = 'family_status',palette='Set1',legend=True)
ax[2].set_title("Family Status Distribution")
ax[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
print('Shows demographic details  what kind of jobs, education levels, and family types customers belong to.')
```

```python

#Housing & Occupation Type
fig, ax = plt.subplots(1, 2, figsize=(20, 8))

sns.countplot(data=df, x='housing_type', ax=ax[0], hue='housing_type',palette='coolwarm',legend=True)
ax[0].set_title("Housing Type Distribution")
ax[0].tick_params(axis='x', rotation=45)

sns.countplot(data=df, x='occupation_type', ax=ax[1], hue='occupation_type',palette='viridis',legend=True)
ax[1].set_title("Occupation Type Distribution")
ax[1].tick_params(axis='x', rotation=50)

plt.tight_layout()
plt.show()

print("Shows where customers live (own house, rented, with parents) and what jobs they have.It helps understand lifestyle and stability.")
```

```python
#Relation Between Income and Credit Card Eligibility
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='target', y='total_income', hue='target',palette='Set2')
plt.title("Income vs Credit Card Eligibility")
plt.show()
print("If the median income of Target = 1 group is higher, income is a strong factor in credit card eligibility.")
```

```python
#Correlation Heatmap

plt.figure(figsize=(8,6))
numeric_df = df.select_dtypes(include=['number'])
sns.heatmap(numeric_df.corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()

print("If theres a strong positive correlation between Num_children and Num_family, it means family size increases with number of children (which is expected).")
print("If Own_property or Own_car correlate with employment status, it shows financial relationships.")
```
